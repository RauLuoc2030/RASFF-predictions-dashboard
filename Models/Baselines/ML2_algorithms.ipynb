# Librerias

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.neural_network import MLPClassifier
from sklearn import tree
from sklearn.metrics import classification_report
```

# Datos entrenamiento

```python
data = pd.read_csv('D:/Visual Studio Code/HK 6/Data Mining/RASFF-predictions-dashboard/Copia csvs antiguos/RASFF_DATA_16092019.csv', sep=';', 
                   header=0, index_col = 0)
df = data
df
```

```python
mask = (df['DATE_CASE'] >= '2004-01-01') & (df['DATE_CASE'] <= '2018-12-31')
df = df.loc[mask]
```

```python
df['DATE_CASE'] =pd.to_datetime(df.DATE_CASE)
df['DATE_CASE'] = df.DATE_CASE.dt.month
```

```python
features = [0,1,6,8]
target = [2]
X = df.iloc[:,features]
y = df.iloc[:,target]
```

```python
X
```

```python
y
```

```python
enc = OneHotEncoder(handle_unknown='ignore', sparse = False)
enc.fit(X.values)
X_one_hot = enc.transform(X.values)
```

# Datos de test

```python
data = pd.read_csv('D:/Visual Studio Code/HK 6/Data Mining/RASFF-predictions-dashboard/Copia csvs antiguos/splited_2019_RASFF_DATA_16092019.csv', sep=';', 
                   header=0, index_col = 0)
test = data
test
```

```python
test['DATE_CASE'] =pd.to_datetime(test.DATE_CASE)
test['DATE_CASE'] = test.DATE_CASE.dt.month
```

```python
X_test = test.iloc[:,features]
y_test = test.iloc[:,target]
```

```python
X_test_OHE = enc.transform(X_test.values)
```

# Decision tree

```python
clf = tree.DecisionTreeClassifier(criterion='gini')
clf = clf.fit(X_one_hot,y)
```

```python
y_pred = clf.predict(X_test_OHE)
```

```python
print(classification_report(y_test,y_pred))
```

```python
clf.score(X_test_OHE, y_test)
```

# Random forest

```python
clf = RandomForestClassifier()
clf = clf.fit(X_one_hot,y)
```

```python
y_pred = clf.predict(X_test_OHE)
```

```python
print(classification_report(y_test,y_pred))
```

```python
clf.score(X_test_OHE, y_test)
```

# Regresión logística

```python
clf = LogisticRegression()
clf = clf.fit(X_one_hot,y)
```

```python
y_pred = clf.predict(X_test_OHE)
```

```python
print(classification_report(y_test,y_pred))
```

```python
clf.score(X_test_OHE, y_test)
```

# MLP básico de sklearn

```python
clf = MLPClassifier(hidden_layer_sizes=(100),max_iter=200)
clf = clf.fit(X_one_hot,y)
y_pred = clf.predict(X_test_OHE)
print(classification_report(y_test,y_pred))
clf.score(X_test_OHE, y_test)
```
